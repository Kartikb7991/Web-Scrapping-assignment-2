```python
driver=webdriver.Chrome(r"C:\Users\karti\OneDrive\Desktop\Data trained\chromedriver.exe")
```

```python
driver.get('https://www.flipkart.com/')
```

```python
close_button=driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']")
close_button.click() 
```

```python
search=driver.find_element_by_xpath("//div[@class='_3OO5Xc']/input")
search.send_keys('sneakers')
```

```python
search_button=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_button.click()
```

```python
Brand=[]
Description=[]
Price=[]
Discount=[]
```

```python
brandnames=driver.find_elements_by_xpath('//div[@class="_2WkVRV"]')[0:100]
for i in brandnames:
    brand=i.text
    Brand.append(brand)
Brand[0:100]
```

```python
desc=driver.find_elements_by_xpath('//div[@class="_2B099V"]/a[1]')
for i in desc:
    description=i.text
    Description.append(description)
```

```python
Description
```

```python
prices=driver.find_elements_by_xpath('//div[@class="_30jeq3"]')[0:100]
for i in prices:
    price=i.text
    Price.append(price)
Price
```

```python
discounts=driver.find_elements_by_xpath('//div[@class="_2B099V"]/a[2]/div/div[3]')
for i in discounts:
    discount=i.text
    Discount.append(discount)
Discount
```

```python
FLIPSNEAK=pd.DataFrame({})
FLIPSNEAK['Brand']=Brand
FLIPSNEAK['Description']=Description
FLIPSNEAK['Price']=Price
FLIPSNEAK['Discount']=Discount
```
