```python
import pandas as pd
import selenium
from selenium import webdriver

from bs4 import BeautifulSoup
import requests

import urllib.request
from urllib.request import urlopen
```

```python
driver=webdriver.Chrome(r"C:\Users\karti\OneDrive\Desktop\Data trained\chromedriver.exe")
```

```python
driver.get('https://www.myntra.com/shoes')
```

```python
price=driver.find_element_by_xpath("//ul[@class='price-list']/li[2]/label/div")
price.click()
```

```python
color=driver.find_element_by_xpath("//li[@class='colour-listItem']/label/div")
color.click()
```

```python
Brand=[]
Description=[]
Price=[]
```

```python
brand=driver.find_elements_by_xpath("//div[@class='product-productMetaInfo']/h3")[:100]
for i in brand:
    Brand.append(i.text)
```

```python
des=driver.find_elements_by_xpath("//div[@class='product-productMetaInfo']/h4[1]")[:100]
for i in des:
    Description.append(i.text)
```

```python
price=driver.find_elements_by_xpath("//div[@class='product-productMetaInfo']/div/span[1]")[:100]
for i in price:
    Price.append(i.text)
```

```python
SHOES=pd.DataFrame({})
SHOES['brand']=Brand
SHOES['des']=Description
SHOES['price']=Price
```
