```python
import pandas as pd
import requests
    
from selenium import webdriver
from bs4 import BeautifulSoup

from selenium.webdriver.chrome.options import Options
chrome_options = Options()
chrome_options.add_argument('--incognito')
```

```python

driver=webdriver.Chrome(r"C:\Users\karti\OneDrive\Desktop\Data trained\chromedriver.exe")
```

```python
driver.get("https://www.flipkart.com/search?q=sunglasses&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY")
```

```python
Brand=[]
Description=[]
Price=[]
Discount=[]
```

```python
brandnames=driver.find_elements_by_xpath('//div[@class="_2WkVRV"]')[0:100]
for i in brandnames:
    brand=i.text
    Brand.append(brand)
Brand[0:100]
```

```python
desc=driver.find_elements_by_xpath('//div[@class="_2B099V"]/a[1]')
for i in desc:
    description=i.text
    Description.append(description)
```

```python
Description
```

```python
prices=driver.find_elements_by_xpath('//div[@class="_30jeq3"]')[0:100]
for i in prices:
    price=i.text
    Price.append(price)
Price
```

```python

discounts=driver.find_elements_by_xpath('//div[@class="_2B099V"]/a[2]/div/div[3]')
for i in discounts:
    discount=i.text
    Discount.append(discount)
Discount
```

```python
FLIP=pd.DataFrame({})
FLIP['Brand']=Brand
FLIP['Description']=Description
FLIP['Price']=Price
FLIP['Discount']=Discount
```
